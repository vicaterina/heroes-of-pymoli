# Purchasing Analysis (Gender)

    *Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
    *Create a summary data frame to hold the results
    *Optional: give the displayed data cleaner formatting
    *Display the summary data frame


```python
# Create a subset of the purchase_data dataframe called get_female_records and gather all data related to Female Gender
# Creating a boolean list to extract only Female="TRUE" rows
get_female_records=purchase_data['Gender']=="Female"
#referencee the get_female_records dataframe to extract only records identified as female
female_data=pd.DataFrame(purchase_data[get_female_records])
#female_data.head()

#Same for male and other
get_male_records=purchase_data['Gender']=="Male"
get_other_records=purchase_data['Gender']=="Other / Non-Disclosed"

male_data=pd.DataFrame(purchase_data[get_male_records])
other_data=pd.DataFrame(purchase_data[get_other_records])

#male_data.head(), other_data.head()
```

```python
#Calculate the average item price for females
female_average_price=round(female_data["Price"].mean(),2)
#female_average_price

male_average_price=round(male_data["Price"].mean(),2)

other_average_price=round(other_data["Price"].mean(),2)
female_average_price, male_average_price, other_average_price
```

```python
#Get total number of purchases for females
total_female_purchases=female_data["Purchase ID"].count() #this is working for female purchase count
#total_female_purchases
#Get total number of purchases for males and other
total_male_purchases=male_data["Purchase ID"].count() #this is working for female purchase count
total_other_purchases=other_data["Purchase ID"].count() #this is working for female purchase count
total_female_purchases,total_male_purchases, total_other_purchases
```

```python
#Sum female purchases Total Purchase Value
total_female_purch=female_data["Price"].sum()
#Sum male and other purchases Total Purchase Value, respectively
total_male_purch=male_data["Price"].sum()
total_other_purch=other_data["Price"].sum()

total_female_purch, total_male_purch, total_other_purch
```

```python
#Take Female sales data and group by individual and sum the individuals purchases and get total for individual
female_purchases=female_data.groupby(by =["SN"]).sum().head()

#Take Male and other sales data and group by individual and sum the individuals purchases and get total for individual
male_purchases=male_data.groupby(by =["SN"]).sum().head()
other_purchases=other_data.groupby(by =["SN"]).sum().head()
```

```python
#Calculate the average purchase per-person/SN
per_fem_purchase=female_data.groupby(by =["SN"]).sum()
per_male_purchase=male_data.groupby(by =["SN"]).sum()
per_other_purchase=other_data.groupby(by =["SN"]).sum()

#Calculate mean from the newly grouped transactions per person 
average_fem_purchase=round(per_fem_purchase["Price"].mean(),2)
average_male_purchase=round(per_male_purchase["Price"].mean(),2)
average_other_purchase=round(per_other_purchase["Price"].mean(),2)

average_fem_purchase, average_male_purchase, average_other_purchase
```
